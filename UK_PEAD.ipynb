```python
column_names = pd.Series(EPS_UK.T.loc[(slice(None),'EPS'),:].index.get_level_values(0).values).str.split().str.get(0)
```

```python
EPS_df = pd.DataFrame(columns=column_names, index=pd.date_range(start='19970331',end='20190930',freq='Q' ))
```

```python
Dates_df = pd.DataFrame(columns=column_names, index=pd.date_range(start='19970331',end='20190930',freq='Q' ))
```

```python
for i in range(0,5*len(column_names),5):
    try:
        Periods = pd.to_datetime(EPS_UK.T.iloc[i:i+5].T.dropna(how='all').iloc[:,0],format='%d/%m/%Y').dt.to_period('Q').dt.end_time
        Periods.name = Periods.name[1]

        try:        
            Dates = pd.to_datetime(EPS_UK.T.iloc[i:i+5].T.dropna(how='all').iloc[:,3],format='%Y%m%d')
            Dates.name = Dates.name[1]

            try:
                EPS_data = EPS_UK.T.iloc[i:i+5].T.dropna(how='all').iloc[:,1].astype('float')
                EPS_data.name = EPS_data.name[1]   
                column_name = EPS_UK.T.iloc[i:i+5].T.dropna(how='all').iloc[:,0].name[0].split()[0]

                try:
                    EPS_data_df = pd.DataFrame(columns = [EPS_data, Periods])
                    Dates_data_df = pd.DataFrame(columns = [Dates, Periods])
                    try:
                        EPS_df.loc[:,column_name] = EPS_data_df.T.reset_index().set_index('Period').EPS
                        Dates_df.loc[:,column_name] = Dates_data_df.T.reset_index().set_index('Period').Date
                    except:
                        print('writing to EPS_df '+column_name)
                except:
                    print('creating EPS_data_df for '+column_name)
            except:
                print('3 '+column_name)
        except:
            print('2 '+column_name)
    except:
        print('1 '+column_name + str(i))
```
